# Pairwise free energies

```python
f = open('free_energies.csv')
pair = f.readlines()
f.close()
```

```python
pair
```

```python
f = open('IC50s.csv')
ic50 = f.readlines()
f.close()
```

```python
ic50_dic = {}
for i in ic50:
    field = i.strip().split(',')
    ic50_dic[field[0]]=float(field[1])
```

```python
def compute_dG(ic50_dic, reference, interest):
    r = ic50_dic[interest]/ic50_dic[reference]
    dG =  0.0019872041*300.00*np.log(r)
    return dG
```

```python
comparison = []
labels = []
for p in pair:
    field = p.strip().split(',')
    exp_dG = compute_dG(ic50_dic,field[0],field[1])
    a = np.array([float(field[2]),float(exp_dG)])
    comparison.append(a)
    labels.append(field[0]+'~'+field[1])
                 
```

```python
c = np.array(comparison)
```

```python
plt.scatter(c[:,0],c[:,1])
```

```python
scipy.stats.pearsonr(c[:,0],c[:,1])
```

```python
n_groups = 15#len(c)

means_men = c[:15,0]


means_women = c[:15,1]


fig, ax = plt.subplots()
fig.set_size_inches((12,12))

index = np.arange(n_groups)
bar_width = 0.35

opacity = 0.4
error_config = {'ecolor': '0.3'}

rects1 = ax.bar(index, means_men, bar_width,
                alpha=opacity, color='b',label='computed')

rects2 = ax.bar(index + bar_width, means_women, bar_width,
                alpha=opacity, color='r',label='experimental')

ax.set_xlabel('ligand pair')
ax.set_ylabel('dG')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(labels[:15], rotation='vertical')
ax.legend()

fig.tight_layout()
plt.show()
```
